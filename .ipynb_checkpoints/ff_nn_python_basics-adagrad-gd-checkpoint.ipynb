```python
#importing the required libraries
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# importing toy dataset with single feature 
data = pd.read_csv('dataset.csv')
```

```python
data #checking the data values
```

```python
# sigmoid neuron implementation with parameters w and b
def sig(w,b,x):
    return(1/(1 + np.exp(-(w*x + b))))
```

```python
# using L2 norm
def error(w,b):
    err = 0.0
    for rows in data.iterrows():
        fx = sig(w,b,rows[1]['X'])
        err += 0.5 * (rows[1]['Y']-fx) ** 2
    return err
```

```python
def grad_b(w,b,x,y):
    fx = sig(w,b,x)
    return (fx - y)*fx*(1-fx)
```

```python
def grad_w(w,b,x,y):
    fx = sig(w,b,x)
    return (fx - y)*fx*(1-fx)*x
```

```python
def adagrad_gradient_descent(): #adaptive learning rates
    w,b,eta,max_epochs = 1,1,10,100
    v_w = 0
    v_b = 0
    epsilon = 1e-8
    for i in range(max_epochs):
        dw, db = 0, 0
        for rows in data.iterrows():
            dw = grad_w(w,b,rows[1]['X'],rows[1]['Y'])
            db = grad_b(w,b,rows[1]['X'],rows[1]['Y'])
        v_w = v_w + dw*dw
        v_b = v_b + db*db
        w = w - (eta/(np.sqrt(v_w+epsilon))) * dw
        b = b - (eta/(np.sqrt(v_b+epsilon))) * db
    print("Loss:",error(w,b))
    print("w:",w)
    print("b:",b)
```

```python
adagrad_gradient_descent()
```
